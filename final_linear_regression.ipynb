```python
import pandas as pd 
import numpy as np
import scipy.stats
import statsmodels.formula.api as smf #to perform anova on categorical columns
house = pd.read_csv('housebuilding.csv') #to read the csv file
house.isnull().sum()                   #to check missing values in our data
```

```python
#threshold is set at 1430 means to retain only those columns which has at least 1430 non NAs from 1460 total values
house.dropna(axis='columns',thresh=1430,inplace=True) 
house.isnull().sum()
#to fill columns with few missing values
house.fillna({'MasVnrType':'BrkFace','MasVnrArea':np.mean(house['MasVnrArea'])},inplace=True)
house.isnull().sum()
```

```python
house.corr() #to find correlation between continous variables
```

```python
#dropping all those columns having zero correlation
house.drop(columns=['Id','MSSubClass','OverallCond','BsmtFinSF2','BsmtUnfSF','LowQualFinSF','BsmtFullBath','BsmtHalfBath','HalfBath','BedroomAbvGr','KitchenAbvGr','EnclosedPorch','3SsnPorch','ScreenPorch','MiscVal','MoSold','YrSold'],inplace=True)
           
```

```python
house.fillna('SBrkr',inplace=True)
```

```python
#to check correlation between categorical independent variables and continuos dependent variable
anova = smf.ols(formula='SalePrice~C(Utilities)', data=house)
result = anova.fit()
print(result.summary())
```

```python
house.shape
```

```python
#performing train test split on our data
from sklearn.model_selection import train_test_split
y = house['SalePrice']
house.drop(columns='SalePrice',inplace=True)

X_train,X_test,y_train,y_test = train_test_split(house,y,test_size=0.2,random_state=24)
```

```python
#label encoding the categorical columns present in our train data
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()

for col in X_train.columns.values:

       if X_train[col].dtypes=='object':
            data=X_train[col].append(X_train[col])
            le.fit(data.values)
            X_train[col]=le.transform(X_train[col])
X_train
```

```python
#label encoding the categorical columns present in our test data
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()

for col in X_test.columns.values:

       if X_test[col].dtypes=='object':
            data=X_test[col].append(X_test[col])
            le.fit(data.values)
            X_test[col]=le.transform(X_test[col])
X_test
```

```python
#applying linear regression machine learning model
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(X_train,y_train)
predicted = model.predict(X_test)
predicted
```

```python
comparision_df = pd.DataFrame({'actual':y_test,'predicted':predicted})
comparision_df
```

```python
#train score or accuracy
model.score(X_train,y_train)
```

```python
#test score or accuracy
model.score(X_test,y_test)
```
